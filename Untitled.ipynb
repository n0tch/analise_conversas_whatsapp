```python
def data_valida(data):
    try:
        datetime.strptime(data, '%m/%d/%y %I:%M %p')
        return True
    except:
        return False
```

```python
def pegar_data_hora(linha):
    try:
        data = linha[0:linha.index(",")].replace(",", "")
        hora = linha[linha.index(" "):linha.index("-")].strip()
        data = (data + ' ' + hora).replace(',', '')
        if(data_valida(data) == True):
            return datetime.strptime(data, '%m/%d/%y %I:%M %p')
        else:
            return None
    except:
        return None
```

```python
def pegar_nome(linha):
    try:
        final = [indice.start() for indice in re.finditer(r":",linha)][1]
        nome = linha[linha.index('-')+1:final].strip()
        if(nome.startswith("Gui")):
            nome = "Guilherme"
        if(nome.startswith("+55 84")):
            nome = "Joelvis"
        
        return nome
    except:
        return None
```

```python
def pegar_texto(linha):
    try:
        inicio = [indice.start() for indice in re.finditer(r":",linha)][1] + 1
        return linha[inicio:].strip()
    except:
        return None
```

```python
def is_alterar_nome(linha):
    if("changed the subject from" in linha):
        return True
    else:
        return False
```

```python
def pegar_nome_alteracao(linha):
    inicio = linha.index("-")
    fim = linha.index(" changed")
    nome = linha[inicio+1:fim].strip()
    if nome == "You":
        nome = "Francisco"
    return nome
```

```python
def pegar_titulo_alteracao(linha):
    inicio = linha.index(" to ")
    titulo = linha[inicio+3:].replace('"','').strip()

    return titulo
```

```python
def cria_dict_titulos(linha):
    dic = {}
    dic['data'] = pegar_data_hora(linha)
    dic['nome'] = pegar_nome_alteracao(linha)
    dic['titulo'] = pegar_titulo_alteracao(linha)
    
    return dic
```

```python
def cria_dict(linha):
    dic = {}
    dic['data'] = pegar_data_hora(linha)
    dic['nome'] = pegar_nome(linha)
    dic['texto'] = pegar_texto(linha)
    
    return dic
```

```python
def criar_lista_dicts(conversa):
    lista = []
    for linha in conversa.readlines():
        lista.append(cria_dict(linha))
    
    return lista
```

```python
def criar_lista_titulo(conversa):
    lista = []
    for linha in conversa.readlines():
        if is_alterar_nome(linha):
            lista.append(cria_dict_titulos(linha))
    
    return lista
```

```python
def count_palavras(palavras):
    elementos = {}
    for palavra in palavras:
        if palavra in elementos.keys():
            elementos[palavra] += 1
        else:
            elementos[palavra] = 1
    
    return elementos
```
